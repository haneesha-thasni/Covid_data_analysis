# **ECDC_COVID_19:DATASET DESCRIPTION**

* dateRep: The date of the reported data

* day: The day of the report (integer, e.g., 1-31).

* month: The month of the report (integer, e.g., 1-12).

* year: The year of the report (integer, e.g., 2020).

* cases: The number of new confirmed cases reported on this date.

* deaths: The number of new deaths reported on this date.

* countriesAndTerritories: The name of the country or territory to which the data belongs.

* geoId: A short geographical identifier (code) for the country or territory.

* countryterritoryCode

### **isin()**

it is used to retreive data is in the dataset.
isin() is a method in Pandas used to filter data. It checks whether each element in a DataFrame or Series is contained in a provided list-like object (like a list, set, or Series). This method is often used for subsetting data based on multiple conditions.

```python
df['countriesAndTerritories'].unique()
```

```python
# Q) filter the dataset to include only rows where the countriesandterritories column is either united_states_of_america or india?
usa_india=df[df['countriesAndTerritories'].isin(['United_States_of_America','India'])]
usa_india
```

```python
# Q) select the rows where geoid is not in a list of excluded codes (excluded_codes=['US','IN','BR'])?
excluded_codes=['US','IN','BR']
```
