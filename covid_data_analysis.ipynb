## Pivot_Table()

*   The pivot function in Pandas is used to reshape or transform data by pivoting the values of one column into multiple columns.

## groupby()

*   The groupby function in pandas is used to split the data into groups based on some criteria, apply a function to each group independently, and then combine the results.
*   Summarized by the "split-apply-combine" pattern.

### Q)  find the maximum number of deaths recorded for each country or territory in the DataFrame

```python
df.groupby('countriesAndTerritories')['deaths'].max()
```

### Q) calculate the average number of cases for each country or territory in the DataFrame

```python
df.groupby('countriesAndTerritories')['cases'].mean()
```

### Q) calculate the total number of deaths for each country or territory for each year in the DataFrame

```python
df.groupby(['countriesAndTerritories','year'])['deaths'].sum()
```

### Q) find the average population data for the year 2018 for each country or territory in the DataFrame

```python
df.groupby('countriesAndTerritories')['popData2018'].mean()
```

### Q)  determine the maximum number of cases and the minimum number of deaths for each country or territory in each year in the DataFrame

```python
df.groupby(['countriesAndTerritories','year']).agg({'cases':max,'deaths':min})
```

### Q) find the minimum, maximum, and average population data for the year 2018 for each country or territory in each year in the DataFrame

```python
df.groupby(['countriesAndTerritories','year'])['popData2018'].agg(['min','max','mean'])
```

```python
df.columns
```

### Q) find the maximum population data for 2018 and the total number of cases for each combination of date, day, month, and year in the DataFrame

```python
df.groupby(['dateRep','day', 'month', 'year']).agg({ 'popData2018':max,'cases':sum})
```

## apply()
*  The apply() method is one of the most common methods of data preprocessing.
* It simplifies applying a function on each row or column in a pandas DataFrame.

### Q) create a new column in the DataFrame that adds 5 to the value in the 'popData2018' column for each row

```python
def add_5(data):
    pop_sum=data['popData2018']+5
    return pop_sum
```

```python
df['pop_sum']=df.apply(lambda x:add_5(x),axis=1)
```

```python
df
```

### Q) create a new column in the DataFrame that adds 5 to each value in the 'deaths' column

```python
df['death_5']=df['deaths'].apply(lambda x:x+5)
```

## cut()
